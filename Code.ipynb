```python
df=df[['Consumer complaint narrative','Sub-product']]  #selecting the columns
df
```

```python
df.dropna(inplace=True)  #removing null values
df
```

```python
# function to group the same class 

def updateClass(text):
    if pd.isna(text):  # Check for missing values
        return "Unknown"
    
    text = text.lower()  # Convert to lowercase for case-insensitive matching
    
    if "debt" in text:
        return "Debt collection"
    elif "loan" in text:
        return "Consumer Loan"
    elif "mortgage" in text:  # Fixed typo
        return "Mortgage"
    else:
        return "Credit reporting, repair, or other"

df["Sub-product"] = df["Sub-product"].apply(updateClass)
```

```python
# Balancing the datset by making the number of samples in each class to the minimum of 116255

from sklearn.utils import resample

target_count = 116255  

 
df_balanced = df.groupby('Sub-product', group_keys=False).apply(lambda x: x.sample(target_count, random_state=42))

df_balanced = df_balanced.sample(frac=1, random_state=42).reset_index(drop=True)

df_balanced.to_csv('balanced_data.csv', index=False)

print(df_balanced['Sub-product'].value_counts())  # Check if all classes are balanced
```

### 2. Text Pre-Processing

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import pandas as pd

# Download stopwords if not already downloaded
# nltk.download('stopwords')
# nltk.download('punkt')

# Precompile stopwords for faster lookup
stop_words = set(stopwords.words('english'))

def preprocess_text_fast(text):
    text = text.lower()  # Convert to lowercase
    text = re.sub(r'\W+', ' ', text)  # Remove special characters
    words = word_tokenize(text)  # Tokenize
    words = [word for word in words if word not in stop_words]  # Remove stopwords
    return " ".join(words)

# Apply preprocessing in a vectorized way
df_balanced['Processed_Text'] = df_balanced['Consumer complaint narrative'].astype(str).map(preprocess_text_fast)

print(df_balanced.head())  # Preview the processed text
```

```python
# Deroping the na value
df_balanced.dropna(inplace=True)
df_balanced.isna().sum()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(max_features=10000, ngram_range=(1,2), stop_words='english', sublinear_tf=True)
X = vectorizer.fit_transform(df_balanced['Processed_Text'])


# Encode target labels
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(df_balanced['Sub-product'])
```

```python
from sklearn.model_selection import train_test_split
# Split dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=22)
```

### 3. Multiclassification models and Comparison of performance


##### A. MultinomialNB

```python

from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report


# Train model
model = MultinomialNB()
model.fit(X_train, y_train)

# Predictions
y_pred = model.predict(X_test)

# Evaluate performance
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred, target_names=le.classes_))
```

```python
# sample prediction
new_complaints = ["I have been charged extra fees on my mortgage account."]
new_complaints_vectorized = vectorizer.transform(new_complaints)

predicted_category = le.inverse_transform(model.predict(new_complaints_vectorized))
print("Predicted Category:", predicted_category[0])
```

##### B. Logistic regression

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
# Sample prediction

new_complaints = ["I have been charged extra fees on my mortgage account."]
new_complaints_vectorized = vectorizer.transform(new_complaints)

predicted_category = le.inverse_transform(model.predict(new_complaints_vectorized))
print("Predicted Category:", predicted_category[0])
```
